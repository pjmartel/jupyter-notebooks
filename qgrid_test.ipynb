```python
import numpy as np
import pandas as pd
import numpy.random as rnd
import ipywidgets as ipyw
import qgrid
pd.options.display.float_format = '{:,.3f}'.format
```

```python
#variables
#Km = 12.3
#Vmax = 86.15566.45
Km = 1
Vmax = 2
smin = 1
smax = 60
step = 5
repl = 3
npoints = 10
std_dev = 0.1
err_type = "absolut"
outl_prob = 0.0
```

```python
# Michaelis-Menten equation
def MM(K,V,s):
  return V*s/(K+s)
```

```python
#generate the data
s = np.linspace(0.2*Km,3.0*Km,npoints)
v = np.zeros((len(s),repl))
for i in range(repl):
  v[:,i] = MM(Km,Vmax,s)
err = rnd.normal(0,std_dev,v.shape)
if err_type == "absolut":
  v = v + err
else:
  v = v * err
from sys import stdout
#np.savetxt(stdout, np.hstack([s.reshape(-1,1),v]),fmt=" %5.2f ")
```

```python
# create a pandas data frame with the data
# also works:
# enz_data.iloc[:,1:4].mean(axis=1)
# enz_data.iloc[:,1:4].std(axis=1)
#
columns = ['[A]']
columns = columns + ['v'+str(i+1) for i in range(repl)]
enz_data = pd.DataFrame(np.zeros((npoints,repl+1)),columns=columns)
enz_data
```

```python
qgrid_widget = qgrid.show_grid(enz_data, show_toolbar=False)
qgrid_widget.layout = ipyw.Layout(width='30%')
qgrid_widget
```
